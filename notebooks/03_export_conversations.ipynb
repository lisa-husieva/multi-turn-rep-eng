```python
import json
import sys
from pathlib import Path

repo_root = Path("..").resolve()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

CONV_ROOT = repo_root / "data" / "conversations"

# ── Config ────────────────────────────────────────────────────────────────────
PAIR_ID = 70          # JBB pair index to export
FRAMEWORKS = {        # folder name → label
    "crescendo_harmful_v2": "crescendo",
    # "actorattack_harmful": "actorattack",
    # "xteaming_harmful":    "xteaming",
}
OUTPUT_PATH = repo_root / "data" / "exports" / f"pair{PAIR_ID:03d}_conversations.json"
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
print(f"Will export pair_id={PAIR_ID} to {OUTPUT_PATH}")
```

```python
conversations = []

for folder_name, label in FRAMEWORKS.items():
    folder = CONV_ROOT / folder_name
    if not folder.exists():
        print(f"Folder not found, skipping: {folder}")
        continue
    pattern = f"*_pair{PAIR_ID:03d}_attempt*.json"
    files = sorted(folder.glob(pattern))
    print(f"{label}: found {len(files)} files for pair_id={PAIR_ID}")
    for fpath in files:
        conv = json.loads(fpath.read_text())
        conv["framework"] = label
        conversations.append(conv)

print(f"\nTotal: {len(conversations)} conversations")
for c in conversations:
    print(f"  attempt={c.get('attempt', '?')}  verdict={c['verdict']}  executed_turns={c['executed_turns']}")
```

```python
OUTPUT_PATH.write_text(json.dumps(conversations, indent=2))
print(f"Saved {len(conversations)} conversations to {OUTPUT_PATH}")
```
